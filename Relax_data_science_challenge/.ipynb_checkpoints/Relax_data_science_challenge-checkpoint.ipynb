### Defining an "adopted user" as a user who has logged into the product on three separate days in at least one seven­day period, identify which factors predict future user adoption.

```python

import pandas as pd

df_engagement = pd.read_csv('takehome_user_engagement.csv')
df_engagement.set_index(pd.DatetimeIndex(df_engagement['time_stamp']), inplace=True)
df_engagement.head()
```

```python
df_users = pd.read_csv('takehome_users.csv',encoding='latin-1')
df_users.head()
```

```python
df_users.shape
```

### Defining an "adopted user" as a user who has logged into the product on three separate days in at least one seven­day period, identify which factors predict future user adoption.

```python
eng_group = df_engagement.groupby('user_id').resample('1W').sum()
```

```python
eng_group.head(10)
```

```python
eng_group = eng_group[eng_group['user_id'].isnull()==False].drop('user_id',axis=1)
eng_group.head(10)
```

```python
import numpy as np
active_users = eng_group['visited'] >=3
active_users = active_users.groupby(level=0).apply(np.sum)
active_users = active_users[active_users!=0]
active_users = pd.DataFrame(active_users,index=active_users.index)
active_users.columns = ['Active Weeks']
active_users.head(5)
```

```python
merged = active_users.merge(df_users, left_index=True, right_on='object_id', how='inner' )
```

```python
merged.head()
```

```python
merged['Active'] = 1
df_users['Active'] = 0
df_users.iloc[list(merged.index),10]=1
model_data = df_users.set_index('object_id').sort_index()
model_data.head(10)
```

```python
# drop unnecessary columns
model_data.drop(['email', 'name','last_session_creation_time'], axis=1, inplace=True)
```

```python
model_data.head()
```

```python
model_data.info()
```

```python
model_data['invited_by_user_id'] = model_data['invited_by_user_id'].fillna(0)
model_data['invited_by_user_id'] =  model_data['invited_by_user_id'].apply(lambda x: 1 if x!=0 else 0)
model_data = pd.get_dummies(data=model_data,columns=['creation_source'])
```

```python
model_data.head()
```

```python
corr_matrix = model_data.corr()
corr_matrix
```

```python
# check the correlation value that's > 0.5 and < 1
corr_matrix = corr_matrix[(np.abs(corr_matrix) >= 0.5) & (np.abs(corr_matrix) < 1)]
corr_matrix
```

```python
#drop date time
model_data.drop('creation_time', axis=1, inplace=True)
```

```python
model_data.head()
```

```python
model_data = pd.get_dummies(data=model_data,columns=['org_id'])
model_data.head()
```

```python
model_data.info()
```

```python
from sklearn.metrics import roc_curve, auc, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
y = model_data.pop('Active')
x = model_data
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.25,random_state=42)
```

```python
max_depth = [1, 2, 3, 4, 5, 6, 7, 8]
param_grid = {'max_depth': max_depth}
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid)
grid_search.fit(x_train, y_train)
print(grid_search.best_params_)
```

```python
clf = DecisionTreeClassifier(max_depth=1).fit(x_train, y_train)
train_score = clf.score(x_train, y_train)

print("train score {}".format(train_score))


test_score = clf.score(x_test, y_test)
print("test score {}".format(test_score))
```

```python
feature_importance = pd.DataFrame(clf.feature_importances_).T
feature_importance.columns = list(x.columns)
feature_importance
```
